## 1. Aggregating Data

Data from four days is combined into one dataframe for analysis

```python
import pandas as pd

#Read in all the files as DF
df1 = pd.read_csv("data/02-21-2018.csv")
df2 = pd.read_csv("data/02-22-2018.csv")
df3 = pd.read_csv("data/02-23-2018.csv")
df4 = pd.read_csv("data/02-28-2018.csv")

#Combine into one file
df = pd.concat([df1, df2, df3, df4])
```

```python
#Save to a compressed file
df.to_pickle("data/aggregate.zip")
```

```python
import pandas as pd

df = pd.read_pickle("data/aggregate.zip")
```

```python
df = df[[]]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df1 = df[df["Dst Port"].between(0, 1000)]
df2 = df[df["Dst Port"] > 1000]
sns.histplot(df1, x="Dst Port", hue="Label")
plt.show()
sns.histplot(df2, x="Dst Port", hue="Label")
plt.show()
```
